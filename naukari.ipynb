```python
skills_series =df['Skills/Description'].dropna().str.lower().str.split().explode().str.strip()
```

```python
top_skills = skills_series.value_counts().head(20)
top_skills_df = top_skills.reset_index()
top_skills_df.columns = ['Skill', 'Count']
print(top_skills_df)
```

```python
df['Job_Role'].value_counts().head(10)
```

```python
# Count jobs per location (top 10)
location_counts = df["Location"].value_counts().head(10)

# Bar plot
plt.figure(figsize=(10, 6))
location_counts.plot(kind='barh', color='lightgreen')
plt.title("Top 10 Locations by Job Count")
plt.xlabel("Number of Jobs")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Count jobs per company (top 10)
company_counts = df["Company"].value_counts().head(10)

# Bar plot
plt.figure(figsize=(10, 6))
company_counts.plot(kind='barh', color='orange')
plt.title("Top 10 Hiring Companies")
plt.xlabel("Number of Jobs")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Count jobs per role (top 10)
role_counts = df["Job_Role"].value_counts().head(10)

# Bar plot
plt.figure(figsize=(10, 6))
role_counts.plot(kind='barh', color='purple')
plt.title("Top 10 Job Roles")
plt.xlabel("Number of Jobs")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
